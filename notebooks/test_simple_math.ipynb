# Test simple math notebook


## Imports
Import wasm based library for rendering svg from DOT located in custom maven repository 

```kotlin
USE {
    repositories {
        maven("https://repsy.io/mvn/yeicor/github-public")
    }

    dependencies {
        implementation("com.github.yeicor:kraphviz:1.0.0")
    }
}
```

### Helpr functions and imports

```kotlin
import org.mikrograd.diff.MLP
import org.mikrograd.diff.Value
import org.mikrograd.diff.minus
import org.mikrograd.diff.div
import org.mikrograd.diff.plus
import org.mikrograd.diff.times
```

```kotlin
import java.io.File
import org.mikrograd.utils.drawDot

fun toDotSvg(value: Value):org.jetbrains.kotlinx.jupyter.api.MimeTypedResult {
    val dotStr = drawDot(value, withGradient = false).render()
    val svg = com.github.yeicor.kraphviz.Kraphviz().render(dotStr)
    return MIME("image/svg+xml" to svg)
}
```

## Multiplication

```kotlin
val a = Value(3.0)
a.label = "a"
val b = Value(2.0)
b.label = "b"
toDotSvg(a * b)
```

```kotlin
val b = Value(2.0)
b.label = "b"
toDotSvg(3 * b)
```

```kotlin
val b = Value(2.0)
b.label = "b"
toDotSvg(3.0 * b)
```
